```python
#Importing the libraries
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels as sm
from scipy import stats
from scipy.stats import sem
from scipy.stats import pearsonr 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from scipy import stats
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error
plt.rcParams['figure.figsize'] = (20.0, 10.0)

#Reading data
data = pd.read_csv('framingham.csv')
print(data.shape)
data['male'] = data['male'].map({1: 'yes', 0: 'no'})
data.head()
```

```python
#check correlation
import seaborn as sns
data_corr = data.corr()
sns.heatmap(data_corr)
```

```python
#locate missing
dataNaN = data[data.glucose.isnull()]
dataNaN
```

```python
#replace NaN value with mean
data['glucose'].fillna(data['glucose'].mean(),inplace=True)
```

```python
sns.set()
sns.histplot(data.age)
plt.show()
```

```python
data.corr()
```

```python
#data appropriate for linear regression
stats.pearsonr(data['sysBP'], data['age'])
```

```python
#Split data
X = data.sysBP
y = data.age
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
#scatter plot
sns.regplot(x = 'sysBP', y = 'age', color = 'blue', data = data, ci= None)
plt.title(" Relationship between Systolic blood pressure and age", fontsize= 15)
plt.show()
```

```python
sns.set()
#plot hist and subplotting
f, axes = plt.subplots(1, 2,figsize=(15, 5))
grid = plt.GridSpec(1, 2, wspace=0.5, hspace=0.3)
plt.subplot(grid[0, 0])
X_train.hist()
plt.title('Systolic blood pressure variance', fontsize= 15)
plt.xlabel('SysBP')
plt.ylabel('Frequency')

plt.subplot(grid[0, 1])
y_train.hist()
plt.title('Age Variance', fontsize= 15)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
